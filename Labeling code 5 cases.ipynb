```python
import pandas as pd
import numpy as np
```

```python
link1 = 'https://drive.google.com/file/d/19Qf4xeuaj_Q8zQmd4USJVTz1oqCSntX-/view?usp=sharing'
path1 = "https://drive.google.com/uc?export=download&id="+ link1.split("/")[-2]
df1 = pd.read_csv(path1)
df1.head()
```

```python
link2 = 'https://drive.google.com/file/d/1k48mi5b3drY94-zqcPjH5Tq-I36Jg5-3/view?usp=sharing'
path2 = "https://drive.google.com/uc?export=download&id="+ link2.split("/")[-2]
df2 = pd.read_csv(path2)
df2.head()
```

```python
link3 = 'https://drive.google.com/file/d/15cXKhfx421I_vSHDhu0RroJJfneMM6h7/view?usp=sharing'
path3 = "https://drive.google.com/uc?export=download&id="+ link3.split("/")[-2]
df3 = pd.read_csv(path3)
df3.head()
```

```python
df = df1.merge(df2, on = 'uuid', how='left')
```

```python
df = df.merge(df3, on = 'ucid', how = 'left')
```

```python
# Delete unecessary columns
df = df.drop(columns = ['is_downgrade','is_upgrade','level','first_login_date_TW','has_teacher_cnt',
                        'has_class_cnt','content_pretty_name','content_kind','subject',
                        'level1_id','level2_id','level3_id','level4_id'])
```

```python
# prompt: in ra số lượng của mỗi giá trị tại cột difficulty

print(df['difficulty'].value_counts())
```

```python
# Column: Difficulty:  unset -> normal
df['difficulty'] = df['difficulty'].replace('unset', 'normal')
```

```python
df = df.drop(columns = ['is_hint_used','user_grade','user_city','gender'])
```

```python
df.head()
```

```python
df['timestamp_TW'] = pd.to_datetime(df['timestamp_TW'])
```

```python
def label_cheating(df):
    def is_valid_group(g):
        return (
            g['uuid'].nunique() >= 2 and
            (g['is_correct'].nunique() == 1)
        )

    valid_groups = df.groupby(['timestamp_TW', 'ucid', 'upid']).filter(is_valid_group)

    df['cheating'] = 0
    df.loc[valid_groups.index, 'cheating'] = 1
    return df
```

```python
def check_cheating(row):
    if row['cheating'] == 1:
        return 1

    if (row['difficulty'] == 'easy' and
        row['total_sec_taken'] < 16.53 and
        row['is_correct'] == True and
        row['used_hint_cnt'] == 0):
        return 1

    if (row['difficulty'] == 'normal' and
        row['total_sec_taken'] < 18.28 and
        row['is_correct'] == True and
        row['used_hint_cnt'] == 0):
        return 1

    if (row['difficulty'] == 'hard' and
        row['total_sec_taken'] < 26.38 and
        row['is_correct'] == True and
        row['used_hint_cnt'] == 0):
        return 1

    if (row['difficulty'] == 'easy' and
        row['total_sec_taken'] > 100.64 and
        row['is_correct'] == True and
        row['total_attempt_cnt'] == 1):
        return 1

    if (row['difficulty'] == 'normal' and
        row['total_sec_taken'] > 131.26 and
        row['is_correct'] == True and
        row['total_attempt_cnt'] == 1):
        return 1

    if (row['difficulty'] == 'hard' and
        row['total_sec_taken'] > 164.86 and
        row['is_correct'] == True and
        row['total_attempt_cnt'] == 1):
        return 1

    return 0
```

```python
label_cheating(df)
df['cheating'] = df.apply(check_cheating, axis=1)
```

```python
print(df['cheating'].value_counts())
```

```python
from sklearn.model_selection import train_test_split

df_train, df_test = train_test_split(
    df,
    test_size=0.2,
    random_state=43
)

print("Percentage of labels in the entire data:")
print(df["cheating"].value_counts(normalize=True))

print("\nPercentage of labels in the training set:")
print(df_train["cheating"].value_counts(normalize=True))

print("\nPercentage of labels in the test set:")
print(df_test["cheating"].value_counts(normalize=True))
```

```python
print("Số dòng tập train:", len(df_train))
print("Số dòng tập test:", len(df_test))
```

```python
# Đếm số lượng nhãn ở tập train
print("Tập train:")
print(df_train["cheating"].value_counts())

# Đếm số lượng nhãn ở tập test
print("\nTập test:")
print(df_test["cheating"].value_counts())
```

```python
def add_label_noise_inplace(df_train, label_col="cheating", noise_level=0.5, seed=None):
    if seed is not None:
        np.random.seed(seed)

    num_noisy = int(len(df_train) * noise_level)  # Number of lines with added noise
    indices = np.random.choice(df_train.index, num_noisy, replace=False)  # Randomly select a row to change

    # Reverse label value (0 → 1, 1 → 0)
    df_train.loc[indices, label_col] = 1 - df_train.loc[indices, label_col]

# Add noise directly to df_train
add_label_noise_inplace(df_train, label_col="cheating", noise_level=0.5)
```
